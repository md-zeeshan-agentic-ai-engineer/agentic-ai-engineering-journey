```python
X = np.array([
    [1, 2],
    [3, 4],
    [5, 6],
    [7, 8]
])

y = np.array([0, 1, 0, 1])
```

```python
print("X:")
print(X)

print("\ny:")
print(y)
```

```python
print("X Shape:", X.shape)
print("y Shape:", y.shape)
```

```python
print(X[0])
print(X[0][1])
print(y[2])
```

```python
X = np.array([
    [180, 80, 40],
    [170, 90, 30],
    [160, 90, 50]
])

y = np.array([1, 0, 1])
```

```python
print("X:")
print(X)

print("\ny:")
print(y)

print("\nX Shape:", X.shape)
print("y Shape:", y.shape)
```

```python
print(X[1])
print(X[2][0])
```
