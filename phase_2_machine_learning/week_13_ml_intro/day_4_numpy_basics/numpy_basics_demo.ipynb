# Arrays
Creating a basic NumPy array.

```python
import numpy as np

a = np.array([1, 2, 3])
print(a)
```

# Shape
Checking the dimensions of the array.

```python
a.shape
```

# Indexing
Accessing specific elements from the array.

```python
print(a[0])
print(a[1:3])
```

# Reshaping
Changing array shap into matrix form.

```python
a = np.array([1, 2, 3, 4])
a.reshape(2, 2)
```

# Math Operations
Performing mathematical operations on arrays.

```python
a = np.array([1, 2, 3])

print(a + 10)
print(a * 2)
print(np.mean(a))
print(np.sum(a))
```

# Task 1: Matrix Creation

```python
print(np.zeros((3,3)))
print(np.ones((2,2)))
print(np.eye(3))
```

# Task 2: Random Arrays

```python
print(np.random.rand(3,3))
print(np.random.randint(1, 10, (2,2)))
```

# Task 3: Array Math

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * b)
print(np.dot(a,b))
```

# Task 4: Reshape + Index

```python
a = np.arange(9)

b = a.reshape(3,3)

print(b)
print(a[0])
print(a[0:5])
```

# MINI CHALLENGE

```python
import numpy as np

a = np.random.randint(1, 100, (3,3))

print("Matrix: ")
print(a)

print("\nSum: ")
print(np.sum(a))

print("\nMean: ")
print(np.mean(a))

print("\nTranspose: ")
print(a.T)

print("\nMax Values: " )
print(np.max(a))
```
