# 2. DataFrame
DataFrame = table like excel sheet.
Example:

```python
import pandas as pd

data = {
    "name": ["MD GHULAM JEELANI", "MD ZEESHAN", "MD MOHIT", "MD MAHIR", "MD HASNAIN"],
    "age": [13, 20, 11, 9, 7],
    "salary": [1000, 2000, 3000, 4000, 6000]
}

df = pd.DataFrame(data)
print(df)
```

# 3. CSV Reading
CSV = dataset file
Example:

```python
import pandas as pd

df = pd.read_csv("employees.csv")

print(df.head())
```

# 4. Columns
single and multiple columns

```python
print(df[["name", "salary"]])
```

# 5. Filtering
Age > 20:
Salary > 40000:

```python
print(df[df["salary"] > 50000])
print(df[df["age"] > 20])
```
