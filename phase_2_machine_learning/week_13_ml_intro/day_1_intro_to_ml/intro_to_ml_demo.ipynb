# Day 1 - Introduction to Machine Learning

## What is Machine Learning?
Machine Learning = Learning patterns from data and using them to make predictions.

## Example:
Input: 1, 2, 3, 4, 5  
Output: 3, 6, 9, 12, 15  

Pattern: Output = 3 × Input

## Key Idea:
We don't hardcode rules in real ML.
Machine learns pattern from data.

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 15]

for i in range(len(x)):
    print(f"{x[i]} -> {y[i]}")
```

```python
def predict(x):
    return 3 * x

predict(50)
```

## Insight

Machine Learning Flow:
Data → Pattern → Model → Prediction

In this example:
Model is manually defined (3x)
In real ML, model learns automatically.
